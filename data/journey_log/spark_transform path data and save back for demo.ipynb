```python
import random
with open('/home/duan/sankey_proj/sankey_notebook/msnbc/msnbc.numbered','r') as f,\
     open('/home/duan/sankey_proj/sankey_notebook/msnbc/msnbc.numbered.new','w') as target_f:
    for line in f:
        aa=line.split(':')
        customer_id = aa[0].strip()
        x=aa[1].split()
        y=[]
        prob = 0.5

        channel_weight=0
        ii = 1
        for i in range(len(x) if len(x) < 4 else 4):
            # print('{}-->1'.format(x[i]))
            ii = i

            ch_id = 1
            # print(channel_dict[x[i]])
            if(int(x[i] )>  7) :
                ch_id = int(x[i]) % 7 + 1
            else:
                ch_id = int(x[i]) 

            y.append(str(ch_id))

            channel_weight = channel_weight + ch_id

        if(ii < 1):
            ii = 1
        if (random.uniform(1, 7) > (channel_weight/ii) ):
            y.append('20')
        else:
            y.append('21')
        
        new_line = str(aa[0]) + " : " + ' '.join(y) + " \n"
        target_f.write(new_line)
    
```
